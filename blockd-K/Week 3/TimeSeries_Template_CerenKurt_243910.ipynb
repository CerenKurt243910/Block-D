### Filling Missing Values

To fill missing values in a time series DataFrame, you can use several approaches depending on the nature of your data. Here are a few common methods using pandas:
-  **Forward Fill:** Propagate the last valid observation forward using the `.ffill()` method. See the documentation [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.ffill.html).
- **Backward Fill:** Propagate the next valid observation backward using the `.bfill()` method. See the documentation  [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.bfill.html#pandas.DataFrame.bfill).
- **Fill with a Specific Value:** You can fill missing values with a specific constant, such as 0, the mean, or median of the column using `.fillna()` method. See the documentation [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html#pandas.DataFrame.fillna).
- **Interpolate Missing Values:** You can use interpolation to estimate missing values based on existing values. There are several methods of interpolation such as linear, polynomial, etc using the `.interpolate()` method. See the documentation [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.interpolate.html#pandas.DataFrame.interpolate).
- **Using Rolling Window:** If your data is noisy, a rolling window approach can be used to fill missing values based on the average of the surrounding values.  

Also there are more advance methods like machine learning imputation in which you can use regression or deep learning models to predict missing values.

Since some workdays are entirely missing from the DataFrame (i.e., they don't even appear as an index), we should **reindex** (see the documenatation [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reindex.html)) first to include all expected workdays before applying any method for filling the missing vaues.

### Calculating Stock Price Increase

Now, we aim to enhance the `data` dataset by adding a new column that indicates whether the closing price of **ASML stock** has increased compared to the previous trading day. By leveraging the `.diff()` method in pandas, we compute the difference between consecutive closing prices. If the difference is positive, we classify the day as an increase (True/1); otherwise, it is classified as no increase (False/0). This new column provides valuable insights for trend analysis, helping identify upward movements in stock prices over time.

```python
# Add a new column 'Price_Increased' that indicates whether the close price increased compared to the previous day
# Use the .diff() method to compute the difference between the current and previous day's close price
# If the difference is positive, it indicates an increase, so the new column will be True, otherwise False
# Put your code here (1 line of ocde)
data["Price_Increased"] = data["Close"].diff() > 0

# Display the first 7 rows to check the results
# Put your code here (1 line of code)
data.head(7)
```

### Calculating Daily Stock Price Percentage Change

To analyze the daily fluctuations in ASML stock prices, we introduce a new column that calculates the percentage change in the closing price compared to the previous trading day. Using the `.shift()` (see the documentation [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.shift.html)) method in pandas, we align each closing price with the previous day's value, allowing us to compute the percentage difference. 

```python
# Calculate the daily percentage change in the closing price
# Use the .shift(1) method to access the previous day's closing price
# Compute the percentage change as ((current close - previous close) / previous close) * 100
# Put your code here (1 line of code)
data["Pct_Change"] = ((data['Close']-data['Close'].shift(1))/data['Close'].shift(1))*100

# Display the first 10 rows to check the calculated percentage change
data.head(10)
```

### Calculating Monthly Stock Price Change

Now we want to analyze monthly close stock price changes by converting dates into monthly periods and computing the price difference between the first and last trading day of each month. Consider the following steps:

- Use `.to_period()` (see the documentation [here](https://pandas.pydata.org/docs/dev/reference/api/pandas.DataFrame.to_period.html#)) method to convert the date index to monthly periods, making it easy to group data by month.
- Use `.groupby()` (see the documentation [here](https://pandas.pydata.org/docs/dev/reference/api/pandas.DataFrame.groupby.html#pandas.DataFrame.groupby)) method to groups stock prices by month.
- Use `.agg()` (see the documentation [here](https://pandas.pydata.org/docs/dev/reference/api/pandas.core.groupby.DataFrameGroupBy.agg.html#pandas.core.groupby.DataFrameGroupBy.agg)) method to find the first and last prices within each month.

```python
# Convert the index (which contains daily dates) to monthly periods
# Use the .to_period('M') method to extract the year and month from the date index
# Store the result in a new column called 'Month'
# Put your code here (1 line of code)
data['Month'] = data.index.to_period('M')


# Display the first few rows to check the new 'Month' column
# Put your code here (1 line of code)
data.head(5)
```

### Stationarity Check for ASML Stock Price

In this section, we apply the Augmented Dickey-Fuller (ADF) test to assess whether the ASML stock price is stationary. The ADF test helps determine if a time series has a unit root, indicating non-stationarity. If the p-value from the test is below a chosen significance level (typically 0.05), we reject the null hypothesis of non-stationarity, suggesting that the ASML stock price is stationary. Conversely, a high p-value indicates that the series may have a trend and is likely non-stationary.

- Use `statsmodels.tsa.stattools.adfuller()` (see the documentation [here](https://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html)) to check the stationarity.

```python
# Import the Augmented Dickey-Fuller test from statsmodels to test for stationarity in the time series.
# Put your code here (1 line of code)
from statsmodels.tsa.stattools import adfuller

# Perform the Augmented Dickey-Fuller (ADF) test on the 'Close' price series to assess stationarity.
# This test evaluates the null hypothesis that the series has a unit root (i.e., it is non-stationary).
# Put your code here (1 line of code)
result = adfuller(ASML_data['Close'])

# Print the p-value from the ADF test results.
# A p-value below 0.05 typically indicates that the null hypothesis can be rejected,
# suggesting the time series is stationary (no unit root).
# Put your code here (1 line of code)
print('P-value:', result[1])
```

### ARIMA Modelling

### Rolling ARIMA Forecasting

Now, let's consider a modified approach called rolling (walk-forward) forecasting, which is the correct way to evaluate an ARIMA model on a test set. Instead of training the model once and making predictions in bulk, we retrain it for each test point using data up to the previous day.

**Step-by-Step Approach for Rolling Forecasting with ARIMA**

1. Split the data into training & test sets
2.  Train the ARIMA model on the training set
3. Iteratively update the model for each test day:
    - Predict only one day ahead
    - Append the real observed value to the training data
    - Retrain ARIMA on the updated dataset
4. Evaluate the model (compare predicted vs actual)

```python
!pip install yfinance
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error


#  2. Split Data (80% Train, 20% Test)
train_data = data.loc[:'2022-12-31'][['Close']]
test_data = data.loc['2023-01-01':'2023-12-31'][['Close']]

#  3. Rolling Forecasting with ARIMA
history = list(train_data['Close'])  # Start with training data
predictions = []

# Loop through each test data point and predict one step ahead
for actual_price in test_data['Close']:
    # Create an ARIMA model instance using the training data ("history")
    # Put your code here (1 line of code)
    model = ARIMA(history, order=(2, 1, 2))  
    
    # Fit the ARIMA model to the training data, "history"
    # Put your code here (1 line of code)
    model_fit = model.fit()
    
    # Use the fitted model to forecast the next time step (the next day).
    # Use 'steps=1' argument meaning we're forecasting only one period ahead.
    # Put your code here (1 line of code)
    forecast = model_fit.forecast(steps=1)
    
    # Extract the predicted value from the forecast result.
    # Since we only predicted one step ahead, we take the first (and only) value.
    # Put your code here (1 line of code)
    predicted_price = forecast[0]
    
    # Store prediction
    predictions.append(predicted_price)
    # Add actual price to history for next iteration
    history.append(actual_price)  

# 4. Evaluate Performance
# Evaluate the model's prediction accuracy using RMSE (Root Mean Squared Error).
# First, compute the Mean Squared Error between actual and predicted closing prices.
# Then, use np.sqrt to take the square root and obtain RMSE.
# RMSE gives a measure of how far predictions are from actual values, in the same unit as the target variable.
# Put your code here (1 line of code)
rmse = np.sqrt(mean_squared_error(test_data['Close'], predictions))

# Print the results
print(f" Root Mean Squared Error (RMSE): {rmse:.2f}")

# 5. Plot Results
plt.figure(figsize=(12,6))
plt.plot(test_data.index, test_data['Close'], label='Actual Price', color='blue')
plt.plot(test_data.index, predictions, label='Predicted Price', color='red', linestyle='dashed')
plt.title("Rolling Forecasting with ARIMA")
plt.legend()
plt.show()
```

### KMeans Clustering

In this subtask, you will apply KMeans clustering to group NASDAQ-100 companies based on their mean daily return and volatility. By using this unsupervised learning method, you will identify clusters of stocks with similar risk and return characteristics. This analysis will help you explore how companies relate to one another in terms of performance and gain insights that can support portfolio diversification or market segmentation.

```python
# Import the KMeans clustering algorithm from scikit-learn for unsupervised learning and clustering tasks
# Put your code here (1 line of code)
from sklearn.cluster import KMeans
```

The following snippet of code implements the Elbow Method to determine the optimal number of clusters for KMeans clustering. The process begins by initializing an empty list, `inertia`, to store the inertia values (the within-cluster sum of squared distances) for different numbers of clusters. The loop iterates over a range of cluster values (`k_range`) from 1 to 9, fitting the KMeans model for each value of k and calculating the corresponding inertia. These inertia values are then appended to the list `inertia`.

After the loop, a plot is created to visualize the relationship between the number of clusters and the inertia values. The x-axis represents the number of clusters, while the y-axis shows the corresponding inertia. The Elbow Method is used to identify the point where adding more clusters does not significantly reduce the inertia, typically forming an "elbow" in the plot. This point is considered the optimal number of clusters for the dataset.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Initialize an empty list to store inertia values for different cluster numbers
# Put your code here (1 line of code)
inertia = []

# Define a range of cluster numbers to test (from 1 to min(9, number of samples))
# Put your code here (1 line of code)
k_range = range(1, min(10, scaled_features.shape[0] + 1))

# Loop over the range of cluster numbers
for k in k_range:
    
    # Initialize the KMeans model with the current number of clusters k
    # Put your code here (1 line of code)
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # Fit the KMeans model to the scaled features
    # Put your code here (1 line of code)
    kmeans.fit(scaled_features)
    
    # Append the inertia value (within-cluster sum of squared distances) to the list
    # You can use kmeans.inertia_
    # Put your code here (1 line of code)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal K')
plt.grid(True)
plt.show()
```

### Anomaly Detection in NASDAQ-100 Stocks: Return and Volatility Analysis

In this subtask, you will explore anomaly detection techniques to identify NASDAQ-100 companies that exhibit unusual return and volatility patterns compared to the rest of the market. These outliers may represent unique investment risks or opportunities. By detecting and analyzing these anomalies, you will gain a deeper understanding of how to recognize irregular behaviors in financial data.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is an unsupervised learning algorithm commonly used for anomaly detection. It works by identifying dense regions of data points, which are grouped together into clusters, and marking sparse regions as outliers. This ability to distinguish between high-density clusters and low-density anomalies makes DBSCAN particularly effective for detecting unusual patterns or outliers in the data. Unlike other methods, DBSCAN does not require specifying the number of clusters in advance and can automatically identify data points that do not fit into any cluster, making it a useful tool for anomaly detection.

**How DBSCAN Detects Anomalies:**

- DBSCAN assigns labels to clusters: 0, 1, 2, etc.
- Points that are too far from dense clusters (don't meet `min_samples` in the `eps` neighborhood) are labeled -1.
- These -1 points are anomalies or noise.

**For stocks, anomalies could be:**

- Stocks with abnormal returns.
- Stocks with extreme volatility.
- Stocks that behave differently from others.

**For this part we use the data without removing the outliers.The data are already availble in summary_df DataFrame.**

This code uses the DBSCAN algorithm for anomaly detection, where points in low-density regions are labeled as anomalies (with label -1). The `eps` and `min_samples` parameters need to be fine-tuned to appropriately identify reasonable outliers based on the dataset's characteristics. The resulting labels are added to the dataframe for further analysis.

```python
# Importing DBSCAN from sklearn.cluster for anomaly detection
# Put your code here (1 line of code)
from sklearn.cluster import DBSCAN

# Set up the DBSCAN model with specified hyperparameters
# eps controls the maximum distance between two points to be considered neighbors
# min_samples defines the minimum number of points to form a dense region (a cluster)
# Put your code here (1 line of code)
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit the DBSCAN model to the anomaly features to identify clusters and outliers
# Put your code here (1 line of code)
dbscan.fit(scaled_features)

# Get cluster labels (-1 means anomaly/noise)
labels = dbscan.labels_

# Add labels to your dataframe (features_anomaly)
summary_df['cluster'] = labels

summary_df.tail(n=10)
```

```python
# Importing the MinMaxScaler from sklearn for feature scaling
# Put your code here (1 line of code)
from sklearn.preprocessing import MinMaxScaler

# Extracting the 'Close' price column from the dataset and reshaping it to a 2D array
# This is necessary because the MinMaxScaler expects a 2D array for scaling
close_prices = data['Close'].values.reshape(-1, 1)
```

```python
# Instantiate the MinMaxScaler with a specified feature range of (0, 1)
# This ensures that the 'Close' price values will be scaled between 0 and 1
# Put your code here (1 line of code) 
scaler = MinMaxScaler(feature_range=(0, 1))

# Fit the scaler to the 'close_prices' data and transform it into the [0, 1] range
# Use .fit_transform() method
# The 'fit' step calculates the necessary scaling parameters (min and max values)
# The 'transform' step then applies this scaling to the 'close_prices' data
# Put your code here (1 line of code) 
scaled_close = scaler.fit_transform(close_prices)
```

```python
# Define the sequence length: how many past days to use for predicting the next day's price
# You will later adjust this value as a hyperparameter for model tuning
# Choose the sequence length
sequence_length = 30 
```

```python
# Initialize empty lists to store the features (X) and target values (y)
X = [] # Features: past 'sequence_length' days' prices
y = [] # Target: the price for the next day (to predict)

# Loop through the scaled_close data starting from 'sequence_length' to the end
for i in range(sequence_length, len(scaled_close)):
    # Append the past 'sequence_length' days of 'Close' prices as the features for prediction (X)
    # 'scaled_close[i-sequence_length:i, 0]' slices the last 'sequence_length' prices
    X.append(scaled_close[i-sequence_length:i, 0])
    
    # Append the next day's 'Close' price as the target value (y)
    # 'scaled_close[i, 0]' refers to the current day's 'Close' price (the target to predict)
    y.append(scaled_close[i, 0])

# Convert X and y lists to numpy arrays for compatibility with machine learning models
X, y = np.array(X), np.array(y)
```

```python
# Reshape X to fit LSTM input (samples, timesteps, features)
# Usen np.reshape function and apply it on X
# Put your code here (1 line of code)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
```

```python
# Define the training set size (e.g., 80% of the data will be used for training)
train_size = int(len(X) * 0.8)

# Split the feature array (X) into training and testing sets
# X_train contains the first 80% of the samples, X_test contains the remaining 20%
X_train, X_test = X[:train_size], X[train_size:]

# Split the target array (y) into training and testing sets
# y_train contains the first 80% of the target values, y_test contains the remaining 20%
y_train, y_test = y[:train_size], y[train_size:]
```

### Using MLP for ASML Stock Price Prediction

In this part, you will use the Multilayer Perceptron (MLP) neural network to predict future stock prices. The MLP model helps capture complex patterns in time series data, offering improvements over traditional methods. You will learn how to preprocess data, train the model, and evaluate its performance in financial forecasting.

```python
# Import the TensorFlow library as 'tf'
# Put your code here (1 line of code)
import tensorflow as tf

# Importing the Sequential model from Keras to build a linear stack of layers for the neural network
# Put your code here (1 line of code)
from tensorflow.keras.models import Sequential

# Importing Dense layer and Dropout to create fully connected layers in the neural network
# Put your code here (1 line of code)
from tensorflow.keras.layers import Dense, Dropout

# Importing Adam optimizer, which is commonly used for training neural networks
# Put your code here (1 line of code)
from tensorflow.keras.optimizers import Adam

# Importing the random module for generating random numbers, which can be used in data manipulation or initialization
# Put your code here (1 line of code)
import random
```

```python
# Set the seed for Python's built-in random number generator to ensure reproducibility
# This allows the random operations to produce the same result every time the code is run
random.seed(42)

# Set the seed for Numpy's random number generator for consistent results across runs
# Ensures that operations like random shuffling or sampling in Numpy remain the same each time
np.random.seed(42)

# Set the seed for TensorFlow's random number generator to ensure reproducibility in model training
# This is especially important for operations like random weight initialization and dropout
tf.random.set_seed(42)
```

```python
# Initialize a Sequential model, which allows you to build a neural network layer by layer in a linear stack
# Put your code here (1 line of code)
model = Sequential()

# Add input layer
# # You are free to number of units and  activation functions based on your own design
model.add(Dense(units=64, activation='relu', input_shape=(X_train.shape[1],)))
# Add a dropout layer to prevent overfitting if needed
None

# Add more hidden layers and dropout layers if needed
# You are free to change the number of layers, units, or activation functions based on your own design
# Put your code here (as many lines as you need)
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=256, activation='relu'))

# last layer
# Add the output layer with 1 neuron for regression (predicting a single continuous value)
# Put your code here (1 line of code)
model.add(Dense(units=1, activation='linear'))

# Compile the model using Adam optimizer and Mean Squared Error loss, suitable for regression problems
# Put your code here (1 line of code)
model.compile(optimizer=Adam(learning_rate=0.0001), loss='mean_squared_error', metrics=['mae'])
```

The following code snippet sets up a model checkpointing mechanism using TensorFlow’s ModelCheckpoint callback to automatically save the best version of a deep learning model during training. It monitors the validation loss (`val_loss`) and saves the model to a file named `best_model_DNN.h5` only when an improvement is detected. The save_best_only=True parameter ensures that only the model with the lowest validation loss is kept, which helps avoid overfitting and preserves the most generalizable version of the model. The `verbose=1` setting prints a message each time a better model is saved, and `mode='min'` tells the callback that lower values of the monitored metric are better.

```python
from tensorflow.keras.callbacks import ModelCheckpoint,EarlyStopping

checkpoint = ModelCheckpoint(
    'best_model_DNN.h5',            # File name to save
    monitor='val_loss',             # What metric to monitor
    verbose=1,                      # Print info when saving
    save_best_only=True,            # Save only the best one (lowest val_loss)
    mode='min'                      # Because lower val_loss is better
)

# EarlyStopping added by me for stopping the training when the model stops improving; time saving
early_stop = EarlyStopping(
    monitor='val_loss',            
    patience=20,                    
    verbose=1,                      
    mode='min'                      
)
```

```python
# Train the model using the training data
# model.fit() handles the training loop and updates the model’s weights
# - epochs: number of times the model will go through the entire training set (you should experiment with different values)
# - batch_size: number of samples processed before the model updates (tune this based on your data and hardware)
# - validation_data: tuple of (X_test, y_test) used to evaluate the model after each epoch
# - callbacks: you can use the 'checkpoint' function defined above to save the best model based on validation loss
# Put your code here (1 line of code)
history = model.fit(
    X_train, y_train,
    epochs=500,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[checkpoint, early_stop]
)

```

The following code snippet generates a plot to visualize the training loss and validation loss over the course of training epochs. It helps to assess the model's performance by showing how well it fits the training data and generalizes to unseen validation data. The plot displays the losses on the y-axis against the number of epochs on the x-axis, allowing you to visually track the model’s learning progress and detect overfitting if the training loss decreases while the validation loss increases.

```python
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange')
plt.title('Training & Validation Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
