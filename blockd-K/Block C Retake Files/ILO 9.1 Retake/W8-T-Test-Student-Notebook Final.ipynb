For now, you just need to know that we need to test whether the data is normally distributed and whether the variances of both samples are equal. Otherwise, our statistical tests would not be valid. What we are going to statistically test is whether there is a statistically significant difference in the mean of a given variable for version A or B. 

### Test Assumptions

```python

df_A.columns = ['Id', 'Start time', 'Completion time', 'Email', 'Name',
                'q1', 'q2', 'q3', 'q4', 'q5']
df_B.columns = ['Id', 'Start time', 'Completion time', 'Email', 'Name',
                'q1', 'q2', 'q3', 'q4', 'q5']
```

```python
import scipy.stats as stats

# Sütun isimlerini sadeleştirdikten sonra: q1, q2, q3, q4, q5
questions = ['q1', 'q2', 'q3', 'q4', 'q5']

# Test for each question
for q in questions:
    print(f"\n Assumption Test Results for: {q}")

    # Normality test
    stat_a, pval_a = stats.shapiro(df_A[q])
    stat_b, pval_b = stats.shapiro(df_B[q])

    print(f" Shapiro-Wilk p-value (Version A): {pval_a:.4f}")
    print(f" Shapiro-Wilk p-value (Version B): {pval_b:.4f}")
    if pval_a > 0.05 and pval_b > 0.05:
        print(" Both distributions are approximately normal.")
    else:
        print(" At least one distribution is not normal.")

    # Variance homogeneity test
    stat_lev, pval_lev = stats.levene(df_A[q], df_B[q])
    print(f" Levene's Test p-value: {pval_lev:.4f}")
    if pval_lev > 0.05:
        print(" Variances are equal.")
    else:
        print(" Variances are not equal.")
```

### Run the T-test

Here you will run the T-test. Based on the assumption check above, decide for each question whether you have to run the bootstrapped version or the regular version of the t-test. Run only the appropriate one, and copy the cells as many times as necessary. 

```python
import numpy as np
from scipy import stats

# Set up random generator for bootstrapped t-test
rng = np.random.default_rng()

# Short column names (your cleaned dataset uses these)
questions_bootstrap = ["q1", "q2", "q3", "q4", "q5"]

for q in questions_bootstrap:
    # Use equal_var=False only for q1 due to unequal variances
    if q == "q1":
        results = stats.ttest_ind(df_A[q], df_B[q], equal_var=False, random_state=rng)
    else:
        results = stats.ttest_ind(df_A[q], df_B[q], random_state=rng)
    
    print(f"\n Bootstrapped t-test results for: {q}")
    print(f"p-value = {results.pvalue:.4f}")
    if results.pvalue < 0.05:
        print(" Significant difference between A and B")
    else:
        print(" No significant difference between A and B")
```

If the results are significant, that means that the means between the two versions of this specific question are different enough to exclude chance for being the cause. So if the B version has a higher/lower average score and is statistically significant, then it works better/worse. If the results are not significant then the changes don't have a real measurable effect so maybe it's not better or maybe the questions don't really measure the effect.
